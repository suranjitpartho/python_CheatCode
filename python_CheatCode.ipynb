```python
# slicing column:
df['hotel'] = df['hotel'].str[2:5]
```

```python
# Merge Two Column
df['combined column'] = df['country'] +'-'+ df['distribution_channel']
```

Conditional Column --

```python
df['arrival_date_day_of_month'].unique()
```

```python
# Custom Formula :
def customcolumn(row):
    if row['arrival_date_day_of_month'] < 8.0:
        return '(1st Week)'
    elif row['arrival_date_day_of_month'] < 15.0:
        return '(2nd Week)'
    elif row['arrival_date_day_of_month'] < 22.0:
        return '(3rd Week)'
    else:
        return '(4th Week)'
df['Week No of the Month'] = df.apply(customcolumn, axis=1)
```

```python
# Table Query: Date Range (YYYY-MM-DD)
df3 = df[(df['reservation_status_date'] > '2015-12-22') & (df['reservation_status_date'] <= '2015-12-31')]
```

Pivot

```python
df_pivot = df.pivot_table(index='deposit_type',
               columns='hotel',
               values='adults',
               aggfunc='sum')

df_pivot = df_pivot.reset_index().rename_axis(None, axis=1)  
```

```python
df.pivot_table(index='deposit_type',
               columns='hotel',
               values=['adults', 'children', 'babies'],
               aggfunc='sum') 
```

```python
df.pivot_table(index='deposit_type',
               columns='hotel',
               values=['adults', 'children', 'babies'],
               aggfunc={'adults': 'sum',
                        'children': 'mean',
                        'babies': 'max'}) 
```

```python
df.pivot_table(index='deposit_type',
               columns='hotel',
               values='adults',
               aggfunc={'sum', 'mean'}) 
```
